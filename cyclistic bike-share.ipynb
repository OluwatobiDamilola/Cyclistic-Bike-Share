```python
#Concatenating all data into a single dataframe
casestudy= pd.concat([june21,july21,aug21,sept21,oct21,nov21,dec21,jan22,feb22,mar22,april22,may22])
```

```python
casestudy.info()
```

```python
casestudy.isna() .sum()
```

```python
casestudy['started_at']= pd.to_datetime(casestudy['started_at'])
casestudy['ended_at']= pd.to_datetime(casestudy['ended_at'])
```

```python
casestudy.isna() .sum()
```

```python
casestudy.info()
```

```python
#deriving Ride_length feature
casestudy['ride_length'] = (casestudy['ended_at'] - casestudy['started_at'])
```

```python
casestudy.head()
```

```python
casestudy['month'] = casestudy.started_at.apply(lambda x: x.month_name())
```

```python
casestudy.head()
```

```python
casestudy['day'] = casestudy.started_at.apply(lambda x: x.day_name())
```

```python
casestudy.columns
```

```python
casestudy.drop('day_of_week',axis=1,inplace=True)
```

```python

casestudy['start_station_name'].fillna(casestudy['start_station_name'].mode()[0], inplace=True)
casestudy['start_station_id'].fillna(casestudy['start_station_id'].mode()[0], inplace=True)
casestudy['end_station_name'].fillna(casestudy['end_station_name'].mode()[0], inplace=True)
casestudy['end_station_id'].fillna(casestudy['end_station_id'].mode()[0], inplace=True)
```

```python
casestudy['end_lat'].fillna(casestudy['end_lat'].mode()[0], inplace=True)
casestudy['end_lng'].fillna(casestudy['end_lng'].mode()[0], inplace=True)
casestudy['ride_length'].fillna(casestudy['ride_length'].mode()[0], inplace=True)
```

```python
casestudy['member_casual'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Customer category')
plt.legend()
plt.savefig('Customer_category.png')
```

```python
casestudy['day'].value_counts().plot(kind='bar',figsize=(6,6))
plt.savefig('Total_number_of_rides_perday.png')
```

```python
import seaborn as sns
sns.countplot(x='day',order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'], data=casestudy, hue='member_casual')
plt.legend(bbox_to_anchor=(1.05,1),loc=2, borderaxespad=0., fontsize=15)
plt.savefig('bike_usage_by_day.png')
```

```python
#bike usage by members
sns.countplot(x='rideable_type', hue='member_casual',data=casestudy, palette='ocean')
plt.savefig('bike_type_usage.png')
```

```python
Insights

The bike-share company has more annual members than casual members, this is because the annual members make more use of the bikes to commute to work and other places during week days.
Casual members make use of the bikes often during weekends for leisure or other personal activities, they contribute the major users of the bikes at weekends.
Both Annual members and Casual members use electric-bikes but more of classic-bikes. while Annual members never made use of docked-bikes except for a portion of the casualmembers.
There is a usual increase in rides during the summer season(June-September) with caual members taking most rides in the beginning of the summer and drastic reduction in rides say during winter seasons(December_February).
```

```python
Recommendation

An attractive discount should be made for membership subscription during the beggining of the summer which include a free or little charge for weekend rides as this would attract casual riders to become annual members.
There should be a monthly give-away or bonanza which would involve casual riders to register as an annual member and get free ride pass for two days during the weekend, since casual riders ride more during the weekend.
Marketing adverts should target all stations and casual riders
```
